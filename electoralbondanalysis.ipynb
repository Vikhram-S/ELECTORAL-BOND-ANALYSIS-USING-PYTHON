```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import plotly.express as px
dfb = pd.read_csv("C:/Users/hpsli/Favorites/Downloads/electroral_bonds_buyer_data.csv")
dfr= pd.read_csv("C:/Users/hpsli/Favorites/Downloads/electroral_bonds_receiver_data.csv")

plt.figure(figsize=(10, 6))

purchase_by_purchaser = dfb.groupby('Purchaser Name')['Denomination'].sum()

top_10_purchasers = purchase_by_purchaser.sort_values(ascending=False).head(30)
top_10_purchasers.plot(kind='bar')

plt.title('Total Purchases by Top 30 Purchasers')
plt.xlabel('Purchaser Name')
plt.ylabel('Total Denomination (in 1000 Crores INR)')
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
dfb = pd.read_csv("C:/Users/hpsli/Favorites/Downloads/electroral_bonds_buyer_data.csv")
dfr= pd.read_csv("C:/Users/hpsli/Favorites/Downloads/electroral_bonds_receiver_data.csv")


# Assuming 'dfr' is your DataFrame and it's already loaded with data

# Calculate the sum of donations for each party and sort them
party_donations = dfr.groupby('Name of the Political Party')['Denomination'].sum().sort_values(ascending=False)

# Select the top twenty-seven parties
top_twenty_seven_parties = party_donations.head(10)

# Define Seaborn color palette to use
colors = sns.color_palette('pastel', n_colors=10)

# Create pie chart
plt.figure(figsize=(10, 8))  # Adjusted for better visibility with more categories
plt.pie(top_twenty_seven_parties, labels=top_twenty_seven_parties.index, colors=colors, autopct='%.0f%%', startangle=0)
plt.title('Electoral Bond Distribution Among Top Ten Political Parties in India')
plt.show()
```

```python
Buyer = dfb.groupby('Purchaser Name')['Denomination'].sum().sort_values(ascending=False)

# Select the top twenty-seven parties
top_ten_buyers = Buyer.head(20)

# Define Seaborn color palette to use
colors = sns.color_palette('pastel', n_colors=20)

# Create pie chart
plt.figure(figsize=(10, 8))  # Adjusted for better visibility with more categories
plt.pie(top_ten_buyers, labels=top_ten_buyers.index, colors=colors, autopct='%.0f%%', startangle=0)
plt.title('Electoral Bond Purchaser Distribution Among Top Twenty Purchaser in India')
plt.show()
```
